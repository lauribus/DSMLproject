```python
#idle time
from datetime import datetime, timedelta

def calculate_idle_time(usage_records):
    total_usage_time = sum((end_time - start_time).total_seconds() for start_time, end_time in usage_records)
    total_time = 365 * 24 * 60 * 60  # Total time in seconds in a year
    
    idle_time = total_time - total_usage_time
    return idle_time

def calculate_bike_idle_times(bike_records):
    bike_idle_times = {}
    
    for bike_id, usage_records in bike_records.items():
        idle_time = calculate_idle_time(usage_records)
        bike_idle_times[bike_id] = idle_time
    
    return bike_idle_times

bike_idle_times = calculate_bike_idle_times(bike_records)

# Print the idle time of each bike
for bike_id, idle_time in bike_idle_times.items():
    print(f"Bike {bike_id}: {idle_time / 3600} hours")
```
